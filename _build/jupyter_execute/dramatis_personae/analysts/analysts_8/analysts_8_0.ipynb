# Step-by-step guide to install all the required packages for creating a Jupyter-book in Woindows 10 using a Linux shell.

## 0. Install Windows Subsystem for Linux (WSL)

- Open the Microsoft Store
- Search for "Linux"
- Click on "Get" for "Ubuntu 20.04 LTS"
- Click on "Launch"
- Create a username and password
- Type the following command in the Ubuntu terminal:

```bash
sudo apt-get update
```

- Type the following command in the Ubuntu terminal:

```bash
sudo apt-get upgrade
```

- Type the following command in the Ubuntu terminal:

```bash
sudo apt-get install python3-pip
```

- Type the following command in the Ubuntu terminal:

```bash
sudo apt-get install git
```

- Type the following command in the Ubuntu terminal:

```bash
sudo apt-get install python3-venv
```

- Type the following command in the Ubuntu terminal:

```bash
sudo apt-get install python3.8-venv
```

## 1. Create a virtual environment

- Type the following command in the Ubuntu terminal:

```bash
python3.8 -m venv myvenv
```

- Type the following command in the Ubuntu terminal:

```bash
source myvenv/bin/activate
```

## 2. Install Jupyter-book

- Type the following command in the Ubuntu terminal:

```bash
pip install -U jupyter-book
```

## 3. Create a Jupyter-book

- Type the following command in the Ubuntu terminal:

```bash

jupyter-book create mybook
```

## 4. Install Jupyter-book dependencies

- Type the following command in the Ubuntu terminal:

```bash
cd mybook
```

- Type the following command in the Ubuntu terminal:

```bash
pip install -r requirements.txt
```

## 5. Build the Jupyter-book

- Type the following command in the Ubuntu terminal:

```bash
jupyter-book build mybook
```

## 6. Serve the Jupyter-book

- Type the following command in the Ubuntu terminal:

```bash

jupyter-book serve mybook
```

## 7. Open the Jupyter-book

- Open a browser

- Type the following URL in the browser:

```bash

http://localhost:4000/
```

## 8. Deactivate the virtual environment

- Type the following command in the Ubuntu terminal:

```bash

deactivate
```

**Early steps**

**Jupyter**

Jupyter Notebooks are a great way to automate your data analysis. They allow you to combine code, text, and images in a single document (as is done throughout this Fena platform). This makes it easy to share your work with others and to reproduce your analysis.

**Workflow**

The workflow for automating your data analysis is as follows:

1. Create a Jupyter Notebook using any of the following methods:
   - Google Colab (unbelievably easy to use)
   - Jupyter lab
   - VS Code

2. Write code in the Jupyter Notebook
   - Jupyter lab and VS Code have a built-in code editor allowing you to write code in Python, R, Julia, and other languages
   - We've developed a workflow that hacks this code editor to allow you to write code in any language you want e.g. Stata
   - Sign up for our [advanced course](https://publichealth.jhu.edu/course/39118) to learn more about this workflow

3. Run the code in the Jupyter Notebook

4. Save the Jupyter Notebook

5. Share the Jupyter Notebook with others
   - Beginners: upload it to Google Colab and sharing a link to the Colab notebook
   - Intermediate: upload it to Jupyter lab and sharing a link to the Jupyter lab notebook
   - Advanced: push it to GitHub pages and share a link to your book (as we've done with this Fena platform)
